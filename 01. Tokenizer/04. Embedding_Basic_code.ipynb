```python
import torch
import torch.nn as nn
```

```python
input_ids = torch.tensor([2, 3, 5, 1])
print(input_ids)
```

```python
vocab_size = 6
output_dim = 3
torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
print(embedding_layer.weight)
```

```python
print(embedding_layer.weight[3])
```

```python
print(embedding_layer(torch.tensor([3])))
```

```python
print(embedding_layer(input_ids))
```

```python
input_index = torch.tensor([2, 5, 4])
print(input_index)

emb_dim = 10
vocab_size = 6

embedding_layer2 = torch.nn.Embedding(vocab_size, emb_dim)
print(embedding_layer2)
print(embedding_layer2.weight)
```

```python
print(embedding_layer2(input_index))
```
